```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
names_1950 = pd.read_csv('names/yob1950.txt', names = ['Name', 'Gender', 'Qty'])
```

```python
pattern = r'^R\w+'
names_1950_R = names_1950[names_1950['Name'].str.contains(pattern)]
top_10_1950_names_R = names_1950_R.groupby(['Name']).sum().sort_values('Qty', ascending = False).head(10)
```

```python
top_10_1950_names_R.plot(kind='pie', y='Qty')
```
